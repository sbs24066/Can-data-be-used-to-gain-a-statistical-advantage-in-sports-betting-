## Average Expected Goals per Season

### I need to split the data into different seasons by using the date feature as the season feature is incorrect data.

```python
Prem_df_XG = Prem_df
```

```python
#season ranges
start_date_2020 = "2020-09-12" 
end_date_2021 = "2021-05-23"

start_date_2021 = "2021-08-13" 
end_date_2022 = "2022-05-22"

start_date_2022 = "2022-08-05" 
end_date_2023 = "2023-05-28"

start_date_2023 = "2023-08-11" 
end_date_2024 = "2024-05-19"
```

```python
#new datasets for each season
PL_Season_20_21 = Prem_df_XG[(Prem_df_XG["date"] >= start_date_2020) & (Prem_df_XG["date"] <= end_date_2021)] 
PL_Season_21_22 = Prem_df_XG[(Prem_df_XG["date"] >= start_date_2021) & (Prem_df_XG["date"] <= end_date_2022)] 
PL_Season_22_23 = Prem_df_XG[(Prem_df_XG["date"] >= start_date_2022) & (Prem_df_XG["date"] <= end_date_2023)] 
PL_Season_23_24 = Prem_df_XG[(Prem_df_XG["date"] >= start_date_2023) & (Prem_df_XG["date"] <= end_date_2024)] 
PL_Season_23_24
```

```python
Prem_df_XG_2024=PL_Season_23_24  
```

```python
COLS_TO_DROP_XG = Prem_df_XG_2024[["time","day","fk"]]
Prem_df_XG_2024.drop(COLS_TO_DROP_XG, axis=1, inplace = True)
Prem_df_XG_2024.head()
```

```python
TEAM_XG_1 = "Manchester City"
TEAM_XG = Prem_df_XG_2024[(Prem_df_XG_2024["team"] == TEAM_XG_1)]
TEAM_XG
```

```python
home = "Home"
TEAM_XG_home = TEAM_XG[(TEAM_XG["venue"] == home)]
TEAM_XG_home.reset_index(drop=True, inplace=True)
TEAM_XG_home
```

```python
#Average amount of goals scored by home team in season
homeGames = TEAM_XG_home["opponent"].count()
totalXG_Home = TEAM_XG_home["xg"].sum()
avgXG_Home = totalXG_Home/homeGames
avgXG_Home
```
